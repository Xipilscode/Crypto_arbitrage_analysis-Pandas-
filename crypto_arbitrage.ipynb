### Step 1: For the bitstamp DataFrame, replace or drop all `NaN`, or missing, values in the DataFrame.

```python
# Convert the Close data type to a float
bitstamp_df.loc[:,"Close"] = bitstamp_df.loc[:,"Close"].astype("float")
bitstamp_df.dtypes
```

```python
# Review the data for duplicate values, and drop them if necessary
bitstamp_df.duplicated().sum()
bitstamp_df = bitstamp_df.drop_duplicates()
```

##### Step 5: Repeat Steps 1–4 for the coinbase DataFrame.

```python
# Repeat Steps 1–4 for the coinbase DataFrame
# For the coinbase DataFrame, replace or drop all NaNs or missing values in the DataFrame
coinbase_df = coinbase.dropna().copy()
coinbase_df.isnull().sum()
```

```python
# Use the str.replace function to remove the dollar sign, $
coinbase_df.loc[:,"Close"] = coinbase_df.loc[:,"Close"].str.replace("$","")
coinbase_df.dtypes
```

#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one. Then use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame, where the spread is greater than zero.



```python
# For the date early in the dataset, measure the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one
arbitrage_spread_early= bitstamp_sliced['Close'].loc[early_date] - coinbase_sliced['Close'].loc[early_date]
arbitrage_spread_middle = bitstamp_sliced['Close'].loc[middle_date] - coinbase_sliced['Close'].loc[middle_date]
arbitrage_spread_late = bitstamp_sliced['Close'].loc[late_date] - coinbase_sliced['Close'].loc[late_date]

# Use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame
print(f"Summary statistics for arbitragespread on {early_date}:")
print(arbitrage_spread_early[arbitrage_spread_early > 0].describe())
print("")
print(f"Summary statistics on {middle_date}:")
print(arbitrage_spread_middle[arbitrage_spread_middle > 0].describe())
print("")
print(f"Summary statistics on {late_date}:")
print(arbitrage_spread_late[arbitrage_spread_late > 0].describe())
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Coinbase from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).
spread_return_early = arbitrage_spread_early[arbitrage_spread_early > 0]/ coinbase_sliced['Close'].loc[early_date]
spread_return_middle = arbitrage_spread_middle[arbitrage_spread_middle > 0]/ coinbase_sliced['Close'].loc[middle_date]
spread_return_late = arbitrage_spread_late[arbitrage_spread_late > 0]/ coinbase_sliced['Close'].loc[late_date]

# Review first 10 rows of the spread return DataFrame
print(f"Spread return on {early_date}:")
display(spread_return_early.head(10))
print("")
print(f"Spread return on {middle_date}:")
display(spread_return_middle.head(10))
print("")
print(f"Spread return on {late_date}:")
display(spread_return_late.head(10))
```

#### 3. For each of the three dates, narrow down your trading opportunities even further. To do so, determine the number of times your trades with positive returns exceed the 1% minimum threshold that you need to cover your costs.

```python
# For the date early in the dataset, determine the number of times your trades with positive returns 
# exceed the 1% minimum threshold (.01) that you need to cover your costs
profitable_trades_early = spread_return_early[spread_return_early > .01]
profitable_trades_middle = spread_return_middle[spread_return_middle > .01]
profitable_trades_late = spread_return_late[spread_return_late > .01]

# Review the first five profitable trades
print(f"Profitable trades on {early_date}:")
display(spread_return_early.head())
print("")
print(f"Profitable trades on {middle_date}:")
display(spread_return_middle.head())
print("")
print(f"Profitable trades on {late_date}:")
display(spread_return_late.head())
```

#### 4. Generate the summary statistics of your spread returns that are greater than 1%. How do the average returns compare among the three dates?

```python
# For the date early in the dataset, generate the summary statistics for the profitable trades
# or you trades where the spread returns are are greater than 1%
print(f"Summary statistics on {early_date}:")
display(profitable_trades_early.describe())
print("")
print(f"Summary statistics on {middle_date}:")
display(profitable_trades_middle.describe())
print("")
print(f"Summary statistics on {late_date}:")
display(profitable_trades_late.describe())
```

#### 5. For each of the three dates, calculate the potential profit, in dollars, per trade. To do so, multiply the spread returns that were greater than 1% by the cost of what was purchased. Make sure to drop any missing values from the resulting DataFrame.

```python
# For the dates in the dataset, calculate the potential profit per trade in dollars 
# Multiply the profitable trades by the cost of the Bitcoin that was purchased
profit_early = profitable_trades_early * coinbase_sliced.loc[early_date]['Close']
profit_middle = profitable_trades_middle * coinbase_sliced.loc[middle_date]['Close']
profit_late = profitable_trades_late * coinbase_sliced.loc[late_date]['Close']

# Drop any missing values from the profit DataFrames
profit_per_trade_early = profit_early.dropna()
profit_per_trade_middle = profit_middle.dropna()
profit_per_trade_late = profit_late.dropna()

# View the first five rows of profit DataFrames
print(f"Profit per trade on: {early_date}")
display(profit_per_trade_early.head())
print("")
print(f"Profit per trade on: {middle_date}")
display(profit_per_trade_middle.head())
print("")
print(f"Profit per trade on: {late_date}")
display(profit_per_trade_late.head())
```
